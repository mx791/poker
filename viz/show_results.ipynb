```python
from theme import theme, make_chart
import matplotlib.pyplot as plt
import numpy as np

THEME = "#C0392B"
theme(theme_color=THEME, back_color="#fff")
```

# 1 - Bot probabiliste

```python
nb = np.array([1, 2, 3, 4, 5])
perf = np.array([0.106000, 0.430000, 0.868000, 1.752000, 2.83400])

def plot(fig):
    plt.bar(nb, perf, color=THEME, label="Performance brute")
    plt.plot(nb, perf/nb, label="Performance normalisée par le nombre de joueur", color="#229954")
    plt.xlabel("Nombre d'adversaires")
    plt.legend(loc=(0, -0.2), ncol=6, frameon=False)

make_chart(
    plot,
    "Performance du bot en fonction du nombre de joueurs",
    "Gain moyen sur 500 simulations - Les adversaires jouent aléatoirement",
    title_color=THEME,
)
```
